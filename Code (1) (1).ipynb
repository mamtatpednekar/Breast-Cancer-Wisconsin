```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)

from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(X_scaled)
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['Principal Component 1','Principal Component 2'])
finalDf = pd.concat([principalDf, dataset[['C']]], axis = 1)
print(finalDf)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1',fontsize = 15)
ax.set_ylabel('Principal Component 2',fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets=[2, 4]
colors=['black','red']
for output, color in zip(targets, colors):
    indicesToKeep = finalDf['C'] == output
    ax.scatter(finalDf.loc[indicesToKeep,'Principal Component 1'],finalDf.loc[indicesToKeep,'Principal Component 2'], c=color, s=50)
ax.legend(targets)
ax.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#decsion tree classification

from sklearn.tree import DecisionTreeClassifier
decision_c = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
decision_c.fit(X_train, y_train)
```

```python
print(decision_c.predict(sc.transform([[4,1,1,3,2,1,3,1,1]])))
```

```python
y_pred = decision_c.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_decision_c = confusion_matrix(y_test, y_pred)
print(cm_decision_c)
accuracy_score(y_test, y_pred)
```
